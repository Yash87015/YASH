## 3. Explain the following terms.1) The difference in np.array, np.asarray and np.asanyarray . 2) The difference between Deep copy and shallow copy

### Ans
**1) the differnce in np.array, np.asarray and np.asanyarray**


**np.array**

→ This is the most common way to create a NumPy array.
→ It creates a new array from any array-like object (e.g., lists, tuples, other arrays).
→ It always copies the input data into a new array.
→ It allows you to specify the data type, memory layout, and other parameters.

**np.asarray()**
→ Similar to np.array(), it creates a NumPy array from an array-like object.
→ This can be more efficient if you're working with large arrays and don't need to modify the original data.

→ *** The key difference is that np.asarray() avoids copying the input data if it is already a NumPy array.***


**np.asanyarray()**
→ This function is even more flexible than np.asarray().

→ ***It converts the input to an ndarray, but it also passes through ndarray subclasses (such as np.matrix) without copying.***

→Use this if you want to work with any array-like object while preserving the specific subclass type.

→ we see each example as well


**2) The difference between Deep copy and shallow copy**

**1)Deep copy**

→Creates a new copy of the object and its data, ensuring that changes to the copy do not affect the original. Deep copies are useful for complex data structures and when data integrity is important. However, they can be slower and use more memory than shallow copies

**2)shallow copy**

→Creates a new object that references the original object's data, so changes to the copy can affect the original. Shallow copies are useful for simple objects like lists, dictionaries, and sets, and can help optimize performance by avoiding duplicating large data structures.

**key difference**

→ The main difference between a deep copy and a shallow copy is that a deep copy creates a new copy of the object and its data, while a shallow copy creates a new object that references the original object's data:




```python
#example and differnce of np.array,npasarry,np.asanyarrry
a=np.array([1,2,3])
b=np.asarray(a)# No copy made
c=np.matrix([[1, 2], [3, 4]])
d=np.asanyarray(c)# No copy made, preserves the matrix type
print(a)
print(b)
print(d)
```

```python
#example of deep copy
import copy
a=c
b=copy.deepcopy(a)
a[0][0]=10
print(a)
print(b)
```

```python
#example of shallow copy
import copy
a=c
b=copy.copy(a)
a[0][0]=10
print(a)
print(b)
```

## 15. Create a NumPy array data containing 1000 samples from a normal distribution. Perform the following tasks using Matplotlib:a) Plot a histogram of the data with 30 bins.b) Overlay a line plot representing the normal distribution's probability density function (PDF).c) Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.d) Set the title of the plot as 'Histogram with PDF Overlay'.

### Ans

```python
#Create a NumPy array data containing 1000 samples from a normal distribution
import numpy as np
data=np.random.normal(size=1000)
data
```

```python
#using Matplotlib: Plot a histogram of the data with 30 bins
import matplotlib.pyplot as plt
plt.hist(data,bins=30)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# Overlay a line plot representing the normal distribution's probability density function (PDF)
#Label the x-axis as 'Value' and the y-axis as 'Frequency/Probability'.# Set the title of the plot as 'Histogram with PDF Overlay'.
import numpy as np
import matplotlib.pyplot as plt
x=np.linspace(-4,4,100)
y=1/(np.sqrt(2*np.pi))*np.exp(-(x**2)/2)
plt.hist(data,bins=30,density=True,alpha=0.5)
plt.plot(x,y,color='red')
plt.xlabel('Value')
plt.ylabel('Frequency/Probability')
plt.title('Histogram with PDF Overlay')
plt.show()
```

## 18. With Bokeh, plot a line chart of a sine wave function, add grid lines, label the axes, and set the title as 'Sine Wave Function'.

### Ans

```python
#create bokeh plot a line chart of sine wave function
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
x=np.linspace(0,4*np.pi,100)
y=np.sin(x)
p=figure(title='Sine Wave Function',x_axis_label='X-axis',y_axis_label='Y-axis')
p.line(x,y,legend_label='Sine Wave',line_width=2)
p.xgrid.grid_line_color='gray'
p.ygrid.grid_line_color='yellow'
show(p)
```

## 19. Using Bokeh, generate a bar chart of randomly generated categorical data, color bars based on their values, add hover tooltips to display exact values, label the axes, and set the title as 'Random Categorical Bar Chart'.

### Ans

```python
#Using Bokeh, generate a bar chart of randomly generated categorical data
import numpy as np
from bokeh.plotting import figure, show
from bokeh.io import output_notebook
output_notebook()
from bokeh.models import ColumnDataSource, HoverTool
from bokeh.transform import factor_cmap
import random
from bokeh.palettes import Spectral6
from bokeh.io import output_file, show
from bokeh.plotting import figure
categories = ['A', 'B', 'C', 'D', 'E']
values = [random.randint(1, 10) for i in range(5)]
source = ColumnDataSource(data=dict(category=categories, value=values))
p = figure(x_range=categories, title="Random Categorical Bar Chart", toolbar_location=None, tools="")
p.vbar(x='category', top='value', width=0.9, source=source, legend_field="category",
       line_color='white', fill_color=factor_cmap('category', palette=Spectral6, factors=categories))
hover = HoverTool(tooltips=[('Category', '@category'), ('Value', '@value')])
p.add_tools(hover)
p.title.text_font_size = '16pt'
p.xaxis.axis_label = 'Category'
p.yaxis.axis_label = 'Value'
p.legend.title = 'Categories'
show(p)
```
