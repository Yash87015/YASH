### Ans

 R-Squared (R² or the coefficient of determination) is a statistical measure in a regression model that determines the proportion of variance in the dependent variable that can be explained by the independent variable. In other words, r-squared shows how well the data fit the regression model (the goodness of fit).

 R-squared formula using Sum of Squares:
 $R^2 = 1 – (RSS/TSS)$


Where,


R2 represents the requrired R Squared value,
RSS represents the residual sum of squares, and
TSS represents the total sum of squares.



→ R-squared can take any values between 0 to 1. Although the statistical measure provides some useful insights regarding the regression model, the user should not rely only on the measure in the assessment of a statistical model. The figure does not disclose information about the causation relationship between the independent and dependent variables.

→ In addition, it does not indicate the correctness of the regression model. Therefore, the user should always draw conclusions about the model by analyzing r-squared together with the other variables in a statistical model.

## Q4.Why do we use Mean Squared Error (MSE)?
### Ans
 Ans Mean Squared Error (MSE) is commonly used in regression and machine learning models as a way to measure the quality of a model’s predictions. Heres why we use MSE:

 1. Quantifies Prediction Accuracy:- MSE measures how far off the predicted values are from the actual values. It is the average of the squared differences between predicted and actual values.


 2. Emphasizes Larger Errors:Squaring the errors means that larger errors are given disproportionately higher weight. This is useful when you want to penalize large deviations more heavily, which can help the model focus on improving those larger errors.
   
 3. Differentiability:MSE is a smooth and differentiable function. This property is crucial for optimization algorithms (e.g., gradient descent) used in machine learning and regression models because it allows the error to be minimized in a continuous manner.

 4. Easily Interpreted and Computed:MSE provides a single numerical value that summarizes the overall performance of a model, making it easy to compare different models or evaluate the improvement of a model during training.It is also computationally simple, making it efficient to calculate, especially for large datasets.

 5. Model Training and Optimization:Minimizing MSE is a common objective in model training. It allows you to adjust the models parameters (such as coefficients in linear regression) to find the best fit.Lower MSE indicates a better fit between the predicted and actual values.

 6. Comparing Models:MSE can be used to compare different models (e.g., different algorithms or model configurations) based on their error rates. A model with a lower MSE is generally considered to perform better.


 Limitations of MSE:Sensitive to Outliers: Since errors are squared, MSE is highly sensitive to outliers. A few large errors can disproportionately increase the MSE, leading to less reliable model performance evaluation.No Direct Interpretability: Unlike metrics like Mean Absolute Error (MAE), MSE doesn’t provide a direct, easily interpretable measure of error because the units are squared. For example, if you’re predicting house prices, MSE’s units will be in squared currency units (e.g., squared dollars), which isn’t intuitive.


 We use Mean Squared Error (MSE) because it quantifies the accuracy of predictions, penalizes larger errors more, is smooth and differentiable for optimization, and provides a simple measure to compare model performance. However, it can be sensitive to outliers and lacks direct interpretability in terms of the original units.

## Q6. How do we check for normality of residuals in linear regression?
### Ans

Checking for the normality of residuals in linear regression is crucial because one of the assumptions of linear regression is that the residuals (errors) should follow a normal distribution. If this assumption is violated, the reliability of hypothesis tests (such as t-tests and F-tests) may be compromised, and the model's predictive power might be affected.

 Here are several common methods to check the normality of residuals:

1. **Histogram**: Plot a histogram of residuals to visually inspect if it resembles a normal distribution (bell curve).

2. **Q-Q Plot**: Create a Q-Q plot to check if the residuals follow a straight line, indicating normality.

3. **Shapiro-Wilk Test**: Perform this test to statistically assess if the residuals are normally distributed (p-value > 0.05 suggests normality).

## Q7 What is multicollinearity, and how does it impact regression ?
### Ans

→ Multicollinearity occurs when two or more independent variables in a regression model are highly correlated with each other. This means that one independent variable can be predicted from another with high accuracy. In other words, the predictors are not truly independent.

→ Impact on Regression:Inflated Standard Errors:Multicollinearity causes inflated standard errors of the regression coefficients, making it difficult to determine the individual impact of each predictor on the dependent variable.This leads to less precise coefficient estimates and can make it harder to assess the significance of predictors.

→ Unreliable Coefficients:When independent variables are highly correlated, the regression model might produce unstable or highly variable coefficients. Small changes in the data can lead to large changes in the coefficient estimates.This reduces the reliability of the model’s predictions and interpretations.

→ Model Overfitting:Multicollinearity can lead to overfitting, where the model fits the training data well but performs poorly on unseen data. This happens because the model may become too sensitive to the noise in the data due to the redundancy between predictors.

→ Reduced Predictive Power:Even though the model might appear to explain a lot of variance in the dependent variable (high R-squared), the presence of multicollinearity can make the model less predictive and generalizable to new data.

→ How to Detect Multicollinearity:Correlation Matrix: Look for high correlation (e.g., above 0.8) between independent variables.Variance Inflation Factor (VIF): A VIF greater than 10 indicates high multicollinearity.

## Q8. What is Mean Absolute Error (MAE)?
### Ans
 Mean Absolute Error (MAE):-Mean Absolute Error (MAE) is a common metric used to evaluate the performance of a regression model by measuring the average magnitude of errors between the predicted and actual values. It tells you how far off, on average, the predictions are from the actual values, without considering the direction of the error (i.e., whether the prediction is higher or lower than the actual value).

### Formula:
\[
MAE = $\frac{1}{n} \sum_{i=1}^{n} |y_i - \hat{y_i}|$
\]
Where:
- $( y_i )$ = Actual values (observed data),
- $( \hat{y_i} )$ = Predicted values (models prediction),
- $( n ) $= Number of observations (data points).

→ Explanation:- Absolute Difference: For each data point, the difference between the actual value and the predicted value is calculated, and the absolute value of this difference is taken (to ignore the direction of the error).Average: The mean of all these absolute errors is then computed, giving the MAE.

→ Interpretation:Lower MAE: Indicates better model performance, meaning the predictions are closer to the actual values.Higher MAE: Indicates poor model performance, meaning the predictions deviate significantly from the actual values.

→ Advantages of MAE:1. simple to Understand: MAE gives an easy-to-understand measure of prediction accuracy.2. No Squaring of Errors: Unlike metrics like Mean Squared Error (MSE), MAE does not square the errors, making it less sensitive to outliers (large errors).

→ Disadvantage1. Less Sensitive to Large Errors: Since it doesnt square the errors, it doesn’t penalize larger errors as heavily as MSE does.2. Does Not Differentiate Between Positive and Negative Errors: MAE doesnt provide insight into whether errors are under-predictions or over-predictions.

## Q16. How is Adjusted R-squared calculated?
### Ans

→  Adjusted R-squared is calculated to provide a more accurate measure of model fit, especially when there are multiple predictors. It adjusts the R-squared value to account for the number of predictors in the model.

### Formula:

$[
\text{Adjusted } R^2 = 1 - \left( \frac{(1 - R^2) \cdot (n - 1)}{n - p - 1} \right)
]$

Where:
- $( R^2 )$ = R-squared value of the model
- $( n )$ = Number of data points (observations)
- $( p )$ = Number of predictors (independent variables) in the model

→ Explanation:Adjusted R-squared adjusts the R-squared value by considering the number of predictors in the model.It penalizes the addition of unnecessary predictors, so if adding more predictors doesn’t improve the model significantly, the adjusted value will decrease.This helps to avoid overfitting by providing a more reliable measure of the model's performance when there are multiple predictors.

## Q19. What is Root Mean Squared Error (RMSE)?
### Ans

→ Root Mean Squared Error (RMSE) is a metric used to measure the difference between predicted values and actual values in a regression model. It quantifies the accuracy of a model by calculating the average magnitude of the errors in the predictions, with larger errors being penalized more heavily due to the squaring of differences.

The formula for RMSE is:

$[
RMSE = \sqrt{\frac{1}{n} \sum_{i=1}^{n} (y_i - \hat{y}_i)^2}
]$

Where:
- $( n )$ is the number of observations (data points).
- $( y_i )$ is the actual value (true value) for the \( i \)-th observation.
- $( \hat{y}_i )$ is the predicted value for the \( i \)-th observation.

→ Root Mean Squared Error (RMSE) is a way to measure how far off a model's predictions are from the actual values. It gives you a single number that shows the average size of the errors.

→ To calculate RMSE:1. Find the difference between each prediction and the actual value.2. Square each difference (this makes big errors more important).3. Take the average of all the squared differences.4. Finally, take the square root of that average.

→ A lower RMSE means better predictions, and a higher RMSE means the model's predictions are more inaccurate.

 ## Q20. Why is pickling considered risky?

###Ans

→ Pickling in Python refers to the process of serializing and deserializing Python objects (i.e., converting objects into a byte stream for storage or transmission and converting them back to their original form). While it can be useful for saving and loading data, pickling can be risky for several reasons:

→ Security Vulnerabilities:When you unpickle (load) data from an untrusted or unknown source, it can execute arbitrary code. This is because pickled data can contain not just objects, but also executable code that can be triggered during unpickling. An attacker could craft a malicious pickle file that, when unpickled, executes harmful code on your system (e.g., deleting files, stealing data).

→ Lack of Compatibility:Pickled data is often Python-specific and can be version-dependent. If you pickle an object in one version of Python and try to unpickle it in a different version, you may run into compatibility issues. This makes it harder to guarantee that pickled objects will work across different Python environments.

→ Data Corruption:If a pickle file is corrupted, it may become unreadable and could cause crashes or errors when trying to unpickle it. This could happen due to file system issues, transmission errors, or disk corruption.

→ Unintended Behavior:Pickled objects might include references to other objects or external resources (such as open files, database connections, etc.). When unpickled, these references may no longer be valid or could lead to unexpected behavior

## Q22. What is heteroscedasticity, and why is it a problem?
### Ans

→ Heteroscedasticity refers to a situation in regression analysis where the variability of the errors (or residuals) is not constant across all levels of the independent variable(s). In simpler terms, it means that the spread or "noise" of the errors changes as the value of the predictor variable(s) increases or decreases.

→ ** Why it's a problem:**
 1. Violates Model Assumptions:Most regression models assume homoscedasticity, which means that the spread of the residuals should be constant across all levels of the independent variable. When this assumption is violated, it can lead to unreliable statistical tests (e.g., t-tests, F-tests) and incorrect inferences.

 2. Inefficient Estimates:Heteroscedasticity can make the model's parameter estimates inefficient. This means the model might not be using the available data effectively, potentially leading to biased or misleading predictions.

 3. Invalid Confidence Intervals:The standard errors of the estimated coefficients may be incorrect, which can make confidence intervals for the model's predictions invalid. This could affect hypothesis testing and lead to misleading conclusions.

## Q23.How can interaction terms enhance a regression model's predictive power?
###  Ans

 Interaction terms in a regression model can enhance predictive power by capturing the combined effects of two or more predictor variables that might not be apparent when considered individually. Essentially, they help account for situations where the impact of one variable on the dependent variable changes depending on the level of another variable.

Heres how they improve predictive power:

1. Capture complex relationships: In real-world data, the relationship between variables may not always be additive. For instance, the effect of education on income might depend on the region you live in. An interaction term between education and region would allow the model to better capture this.

2. Improve accuracy: By incorporating interaction terms, the model can account for nuances in the data that were previously missed, leading to more accurate predictions.

3. Avoid bias: If interaction effects are omitted when they exist, the model might incorrectly attribute the effect to the main variables alone, which can lead to biased estimates. Including interaction terms ensures that these relationships are properly modeled.

4. Model more sophisticated relationships: Some patterns in the data may only emerge when two or more predictors combine. Interaction terms help expose these patterns, improving the model’s overall understanding and prediction of the dependent variable.

→ For example, imagine you're predicting someone's salary based on their years of experience and education level. The effect of experience on salary might be different depending on whether the person has a high school diploma or a college degree.

By adding an interaction term between experience and education, the model can capture this change in effect. Without the interaction term, the model might only show how experience and education separately affect salary, but not how they work together.

In short, interaction terms let the model capture more complex relationships between variables, improving its ability to predict outcomes accurately.

## Q11.. Write a Python script that fits polynomial regression models of different degrees to a synthetic dataset and compares their performance.

### Ans

```python
#sythetic data
np.random.seed(0)
x = np.random.rand(100,1) * 6
y = 2 + 3 * x - 1.5*x**2 + 0.5 * x**3 + np.random.randn(100,1)

#train test split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)

#compare degree with r2 score
def compare_degree(x_train,x_test,y_train,y_test):


    for degree in range(1,6):

      poly_features=PolynomialFeatures(degree=degree)
      x_train_poly=poly_features.fit_transform(x_train)
      x_test_poly=poly_features.transform(x_test)
      model=LinearRegression()
      model.fit(x_train_poly,y_train)
      y_pred=model.predict(x_test_poly)
      r2=r2_score(y_test,y_pred)
      print(f'degree {degree} r2 score {r2}')

#compare
compare_degree(x_train,x_test,y_train,y_test)
```
