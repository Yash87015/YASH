## 5.  Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap

### Ans

```python
#generate a synthetic dataset with correlated feature
import numpy as np
correlation_matrix = np.array([[1.0, 0.7, 0.3],
                             [0.7, 1.0, 0.5],
                             [0.3, 0.5, 1.0]])
num_samples = 11
data = np.random.multivariate_normal(mean=[0, 0, 0], cov=correlation_matrix, size=num_samples)

import pandas as pd
df = pd.DataFrame(data, columns=['feature1', 'feature2', 'feature3'])
```

## 4. Using the given x and y data, generate a 3D surface plot to visualize the function Z = \sin(\sqrt{x^2 + y^2})
## x = np.linspace(-5, 5, 100)

##x, y = np.meshgrid(x, y)
##z = np.sin(np.sqrt(x**2 + y**2))

##    'Y': y.flatten(),

##}
##df = pd.DataFrame(data)

### Ans

```python
#generating  data
x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data={'x': x.flatten(),
'y': y.flatten(),
'z': z.flatten()}
df = pd.DataFrame(data)
```

```python
#import libray
import plotly.express as px
import plotly.graph_objects as go
import numpy as np
```

```python
#create 3D surface plot
surface=go.Surface(x=x,y=y,z=z)
fig=go.Figure(surface)
fig.update_layout(title='3D surface plot of Z = sin(sqrt(x^2 + y^2))', scene=dict(
                       xaxis_title='x',
                       yaxis_title='y',
                       zaxis_title='z'))
fig.show()
```

## 1.Create a Bokeh plot displaying a sine wave. Set x-values from 0 to 10 and y-values as the sine of x.

### Ans

```python
#creating data
x=np.linspace(0,10,10)
y=np.sin(x)
```

```python
#creat bokeh plot and display sine wave
import bokeh.io
import bokeh.plotting
bokeh.io.output_notebook()
from bokeh.plotting import figure, show
p=figure(title='sine wave',x_axis_label='x',y_axis_label='y=sin(x)',)
p.line(x,y,legend_label='sine',line_width=2)
show(p)
```

## 2.Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the markers based on the 'sizes' and 'colors' columns.

### Ans

```python
#create random generated x and y value
from bokeh.models import widgets
from bokeh.models import ColumnDataSource
np.random.seed(3)
x=np.random.rand(10)
y=np.random.rand(10)
```

```python
#create a bokeh scatter plot using x,y and use different size and colours for the marker based on the sizes and colours columns
sizes=np.random.rand(10)*30
colors=['red','green','blue','yellow','orange','purple','pink','brown','gray','black']
a=ColumnDataSource(data=dict(x=x,y=y,sizes=sizes,colors=colors))
p=figure(title='scatter plot',x_axis_label='x',y_axis_label='y')
p.scatter('x','y',source=a,size='sizes',color='colors',fill_alpha=10)
show(p)
```

## 3. Generate a Bokeh bar chart representing the counts of different fruits using the following dataset.
## fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
## counts = [20, 25, 30, 35]

### Ans




```python
#given data
fruits = ['Apples', 'Oranges', 'Bananas', 'Pears']
counts = [20, 25, 30, 35]
```

```python
#generate bar chart representing the counts of different fruits
from bokeh.models import ColumnDataSource
a=ColumnDataSource(data=dict(fruits=fruits,counts=counts))
p=figure(x_range=fruits,title='bar chart',x_axis_label='fruits',y_axis_label='counts')
p.vbar(x='fruits',top='counts',width=0.5,source=a)
show(p)
```

## 4. Create a Bokeh histogram to visualize the distribution of the given data.
## data_hist = np.random.randn(1000)
## hist, edges = np.histogram(data_hist, bins=30)

### Ans

```python
#given data
data_hist = np.random.randn(1000)
hist, edges = np.histogram(data_hist, bins=30)
```

```python
#create bokeh histogram
from bokeh.models import ColumnDataSource
from bokeh.plotting import figure, output_file, show
p=figure(title='histogram',x_axis_label='data',y_axis_label='counts')
p.quad(top=hist,bottom=0,left=edges[:-1],right=edges[1:],fill_color='blue',line_color='black')
show(p)
```
