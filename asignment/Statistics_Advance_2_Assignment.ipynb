**Definition:-**
A z-score is a statistical measure that describes how far a raw score is from the mean, measured in standard deviation units. A positive z-score means the value is above the mean, and a negative z-score means the value is below the mean.

→A z-score of 0 means the data point's score is identical to the mean score.

→A positive z-score indicates the score is above the mean, while a negative z-score indicates it is below the mean.

→In investing and trading, z-scores can be used to help determine volatility.

→ In process control applications, the Z value provides an assessment of the degree to which a process is operating off-target.



→ The z-statistic is a measure that describes the position of a raw score in relation to the mean of a group of scores, expressed in terms of standard deviations. It is calculated using the formula:

$z = \frac{(X - \mu)}{\sigma}$

Where:

 $X$= raw score

 μ= population mean

 σ = population standard deviation

**Relationship to the standard normal distribution:-**
A standard normal distribution (SND) is a normal distribution with a mean of 0 and a standard deviation of 1. Z-scores are used to convert a normal distribution to a standard normal distribution so that probabilities can be found using a z-score table.

→ The z-statistic follows a standard normal distribution, which is a normal distribution with a mean of 0 and a standard deviation of 1.

**Use in hypothesis testing:-**
A z-test is used to test a hypothesis when the population variance is known, or when the sample size is large enough to estimate the population variance. The null hypothesis can be rejected if the z-test statistic is statistically significant when compared with the critical value.

→ Z-tests are often used when the variable of interest is continuous, such as age, weight, or temperature

→ In hypothesis testing, the z-statistic is used to determine whether to reject the null hypothesis. A z-score indicates how many standard deviations an element is from the mean. ## Question 5: What are the key properties of the Poisson distribution, and when is it appropriate to use this distribution?

### Ans
**Poisson distribution:-**the Poisson distribution is a discreate probability distribution that describe the no of events that occur within a fixed interval of time or space given a known average rate of occurrence.

**The key properties of the Poisson distribution are:-** events occur independently, at a constant average rate, and the probability of an event occurring in a small time interval is proportional to the length of that interval; it's appropriate to use when you know the average rate of events occurring within a fixed time period, and the events happen independently of each other, like the number of phone calls received in an hour at a call centre, where each call is considered independent of the others.

**Key points about the Poisson distribution:**
1. Independent events:
The occurrence of one event does not influence the probability of another event happening.

1. Constant average rate:
The average rate of events occurring within a given time frame is known and remains constant.

1. Discrete distribution:
The Poisson distribution models the number of times an event occurs, which can only be a non-negative integer.


4. Parameter: It is characterized by a single parameter \( \lambda \) (lambda), which represents the average rate (mean) of occurrences in the given interval.

3. Probability Mass Function (PMF): The PMF of a Poisson distribution is given by:
   $[
   P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
   ]$
   where \( k \) is the number of occurrences, \( e \) is the base of the natural logarithm, and \( k \) is a non-negative integer (0, 1, 2, ...).
4. Mean and Variance: Both the mean and variance of a Poisson distribution are equal to \( \lambda \).

5. Memoryless Property: The number of events occurring in non-overlapping intervals is independent.

**When to use the Poisson distribution:-**
1. Counting rare events: When you are counting the number of times a relatively rare event occurs within a specific time period or area.

1. Known average rate: If you know the average rate at which events occur, like the average number of customers arriving at a store per hour.

1. Independent occurrences: When the events are considered to be independent of each other.


**Examples of situations where a Poisson distribution might be used:**
→ The number of car accidents on a highway per day.

→ The number of radioactive particles detected in a given time interval.

→ The number of customer arrivals at a shop during a specific time period.

→ The number of typos on a page of text.


In summary, the Poisson distribution is valuable for modeling the number of events in fixed intervals when those events are rare and occur independently.

**Probability density function(PDF):-**The Probability Density Function(PDF) defines the probability function representing the density of a continuous random variable lying between a specific range of values. In other words, the probability density function produces the likelihood of values of the continuous random variable.

→ A PDF is used when dealing with continuous random variables, where the variable can take on any value within a given range, and the probability of a specific value is essentially zero.

→ To find the probability of a continuous variable falling within a certain range using a PDF, you need to integrate the PDF over that range.

→ The area under the curve of the PDF over a given interval represents the probability that the random variable falls within that interval.

→ The total area under the PDF curve is equal to 1.

**Properties**:
1. $( f(x) \geq 0 ) for all ( x ).$
2. The total area under the curve of the PDF is equal to 1:
   $[
   \int_{-\infty}^{\infty} f(x) \, dx = 1
   ]$

**Example**:
Consider a continuous random variable that follows a normal distribution with mean (μ = 0 ) and standard deviation (σ = 1 ).

- **Calculating PDF**:
The PDF of a normal distribution is given by:
$[
f(x) = \frac{1}{\sigma \sqrt{2\pi}} e^{-\frac{(x - \mu)^2}{2\sigma^2}}
]$

For our example:
$[
f(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}
]$

- **Calculating Probability**:
To find the probability that \( X \) falls within the interval \( [a, b] \), you would integrate the PDF:
$[
P(a < X < b) = \int_{a}^{b} f(x) \, dx
]$

## Question 7: Explain the Central Limit Theorem (CLT) with example.

### Ans
**Central limit theorem(CLT):-**The central limit theorem (CLT) states that the distribution of a sample will approximate a normal distribution (i.e., a bell curve) as the sample size becomes larger, regardless of the population's actual distribution shape.

→ Put another way, CLT is a statistical premise that, given a sufficiently large sample size from a population with a finite level of variance, the mean of all sampled variables from the same population will be approximately equal to the mean of the whole population. Furthermore, these samples will approximate a normal distribution, with their variances being approximately equal to the variance of the population as the sample size gets larger, according to the law of large numbers.

**Key Takeaways**
1. The central limit theorem (CLT) states that the distribution of sample means approximates a normal distribution as the sample size gets larger, regardless of the population's distribution.
2. A sufficiently large sample size can predict the characteristics of a population more accurately.
3. Sample sizes equal to or greater than 30 are often considered sufficient for the CLT to hold.
4. A key aspect of CLT is that the average of the sample means and standard deviations will equal the population mean and standard deviation.
5. CLT is useful in finance and investing when analyzing a large collection of securities to estimate portfolio distributions and traits for returns, risk, and correlation.

3. Population Mean and Variance:
   - The mean of the sampling distribution of the sample means will equal the population mean ( μ ).
   - The variance of the sampling distribution will equal the population variance ( σ^2 ) divided by the sample size ($ n)$:
   $[
   \sigma^2_{\bar{x}} = \frac{\sigma^2}{n}
   $]

4. Independence: The samples should be independent. This is often achieved through random sampling.

5. Application to Inference: The CLT justifies the use of normal distribution methods for hypothesis testing and confidence intervals, even when the population distribution is unknown or non-normal.

6. Practical Implications: The CLT enables statisticians to make inferences about population parameters based on sample statistics, facilitating decision-making in various fields such as economics, psychology, and medicine.

**Conditions of the central limit theorem:-**
The central limit theorem states that the sampling distribution of the mean will always follow a normal distribution under the following conditions:

→ The sample size is sufficiently large. This condition is usually met if the sample size is n ≥ 30.

→ The samples are independent and identically distributed (i.i.d.) random variables. This condition is usually met if the sampling is random.

→ The population’s distribution has finite variance. Central limit theorem doesn’t apply to distributions with infinite variance, such as the Cauchy distribution. Most distributions have finite variance.

## Question 8: Compare z-scores and t-scores. When should you use a z-score, and when should a t-score be applied instead?

### Ans


→ Here's a comparison between z-scores and t-scores, along with guidelines for when to use each:

#### Z-Scores vs. T-Scores

| **Aspect**               | **Z-Score**                                | **T-Score**                              |
|--------------------------|--------------------------------------------|------------------------------------------|
| **Definition**           | Measures how many standard deviations a data point is from the population mean. | Measures how many standard deviations a data point is from the sample mean, accounting for sample size. |
| **Distribution**         | Based on the standard normal distribution (N(0, 1)). | Based on the t-distribution, which is wider and has heavier tails than the normal distribution. |
| **Sample Size**          | Generally used when the sample size is large (usually \( n > 30 \). | Used when the sample size is small (usually \( n < 30 \). |
| **Population Variance**  | Used when the population standard deviation ( σ \) is known. | Used when the population standard deviation is unknown and estimated from the sample. |
| **Applicability**        | Commonly used for larger samples and in situations where the central limit theorem applies. | More appropriate for smaller samples, especially when the sample mean is used to estimate the population mean. |

#### When to Use Each

1. **Use a Z-Score When**:
   - The sample size is large ( n > 30 ).
   - The population standard deviation ( σ ) is known.
   - You are working with normally distributed data or the central limit theorem applies (i.e., the sample means will be normally distributed).

2. **Use a T-Score When**:
   - The sample size is small (n < 30).
   - The population standard deviation is unknown and must be estimated from the sample.
   - The data may not be perfectly normally distributed, but the t-distribution accounts for this uncertainty, particularly with smaller samples.

## Question 9: Given a sample mean of 105, a population mean of 100, a standard deviation of 15, and a sample size of 25, calculate the z-score and p-value. Based on a significance level of 0.05, do you reject or fail to reject the null hypothesis? Task: Write Python code to calculate the z-score and p-value for the given data.Objective: Apply the formula for the z-score and interpret the p-value for hypothesis testin

### Ans

 Given Data
- Sample Mean $(\bar{X})$: 105
- Population Mean ( μ ): 100
- Standard Deviation ( σ ): 15
- Sample Size $( n )$: 25
- Significance Level $( \alpha )$: 0.05

**Step 1: Calculate the Z-Score**

The Z-score formula is given by:

$[
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}}
]$

Substituting in the given values:

$[
Z = \frac{105 - 100}{15 / \sqrt{25}} = \frac{5}{15 / 5} = \frac{5}{3} \approx 1.67
]$

**Step 2: Calculate the P-Value**



Using a Z-table, we find:

- The p-value corresponding to ( Z = 1.67 ) is approximately ( 0.9525 ).
  


$[
\text{P-value} = 1 - 0.9525 \approx 0.0475
]$

**Step 3: Make a Decision**

**Decision Rule**:
- If the p-value is less than or equal to the significance level $( \alpha = 0.05 )$, we reject the null hypothesis.
- If the p-value is greater than the significance level, we fail to reject the null hypothesis.

In this case, the p-value \( 0.0475 \) is less than \( 0.05 \).

**Interpretation of the Results**
Z-Score: This tells us how many standard deviations the sample mean is from the population mean.

P-Value: This indicates the probability of observing a sample mean at least as extreme as the one obtained if the null hypothesis is true.

Decision: If the p-value is less than 0.05, we reject the null hypothesis, suggesting that the sample provides enough evidence to conclude that the population mean is different from 100. If it’s greater than 0.05, we fail to reject the null hypothesis.

**Conclusion**

Based on the significance level of 0.05, we **reject the null hypothesis**. This suggests that there is statistically significant evidence to support that the sample mean (105) is significantly different from the population mean (100).

```python
#given data
sample_mean = 105
population_mean = 100
standard_deviation = 15
sample_size = 25
significance_level = 0.05

# Calculate the Z-score
z_score = (sample_mean - population_mean) / (standard_deviation / (sample_size ** 0.5))

# Calculate the p-value
import scipy.stats as stats
p_value = 1 - stats.norm.cdf(z_score)

# Make a decision
if p_value <= significance_level:
    decision = "Reject the null hypothesis"
else:
    decision = "Fail to reject the null hypothesis"

print(f"Z-Score: {z_score}")
print(f"P-Value: {p_value}")
print(f"Decision: {decision}")
```

### Ans


```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Parameters
n_trials = 10
p_success = 0.6
num_samples = 1000

# Simulate binomial distribution
binomial_data = np.random.binomial(n_trials, p_success, num_samples)

# plot the distribution
plt.figure(figsize=(10, 6))
sns.histplot(binomial_data, bins=range(n_trials + 1), kde=True, color='orange')
plt.title('Binomial Distribution')
plt.xlabel('Number of Successes')
plt.ylabel('Frequency')
plt.show()

# Calculate expected mean and variance
expected_mean = n_trials * p_success
expected_variance = n_trials * p_success * (1 - p_success)

print(f"Expected Mean: {expected_mean}")
print(f"Expected Variance: {expected_variance}")



```
