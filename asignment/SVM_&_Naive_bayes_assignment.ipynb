# Theoretical



## Q1. What is a Support Vector Machine (SVM)?
-  A Support Vector Machine (SVM) is a supervised machine learning algorithm that aims to find the optimal hyperplane to separate data points into different classes by maximizing the margin between the closest data points from each class, called "support vectors", making it particularly effective for complex, high-dimensional data classifications.

- A Support Vector Machine (SVM) is a machine learning model used for classification and regression tasks. It tries to find the best boundary (hyperplane) that separates data points into different classes

## Q2. What is the difference between Hard Margin and Soft Margin SVM?
- Hard Margin SVM:
Attempts to find a hyperplane that perfectly separates all data points without any misclassifications, assuming the data is linearly separable, which is often unrealistic in real-world scenarios.
- Soft Margin SVM:
Allows for some misclassifications by introducing "slack variables", making it more robust to noisy data and better suited for non-linearly separable datasets.

## Q3. What is the mathematical intuition behind SVM?
- The mathematical intuition behind Support Vector Machines (SVM) is to find the optimal hyperplane in a multi-dimensional space that best separates data points into different classes by maximizing the margin between them, essentially aiming to create the widest possible gap between the decision boundary and the closest data points from each class, known as "support vectors" - this is achieved by solving a constrained optimization problem with the goal of maximizing the margin while minimizing misclassifications.

- The core idea is to maximize the margin between the hyperplane and the closest data points from each class, which is achieved by minimizing the norm of the weight vector (w) while ensuring that the data points are classified correctly with a margin constraint.

## Q4. What is the role of Lagrange Multipliers in SVM?
- In Support Vector Machines (SVM), Lagrange multipliers act as weights that enable the optimization process to find the best separating hyperplane by balancing the objective of maximizing the margin between classes while still satisfying the constraints imposed by the data points, particularly the "support vectors" which are the most crucial data points near the decision boundary; essentially, they allow the algorithm to efficiently solve the constrained optimization problem involved in SVM classification by assigning importance to each data point based on its proximity to the hyperplane.

- Lagrange multipliers are used in the SVM optimization problem to incorporate the constraints of the margin maximization, allowing for the calculation of the optimal hyperplane while considering the "support vectors" that are most crucial for the classification.

## Q5. What are Support Vectors in SVM?
- Support vectors are the data points that lie closest to the hyperplane and significantly influence the decision boundary, as they are the key elements in determining the margin.

## Q6. What is a Support Vector Classifier (SVC)?
- A Support Vector Classifier (SVC) is a supervised machine learning algorithm specifically designed for classification tasks, where it aims to find the optimal hyperplane that best separates data points into different classes by maximizing the margin between them, essentially identifying the "best" decision boundary to distinguish between categories; it is a type of Support Vector Machine (SVM) focused on classification problems.


## Q7. What is a Support Vector Regressor (SVR)?
- A Support Vector Regression (SVR) is a machine learning algorithm, considered a variant of Support Vector Machines (SVM), used for regression tasks where it aims to find a hyperplane that best fits a dataset by maximizing the margin between the predicted values and actual data points, while primarily focusing on the most critical data points called "support vectors" to achieve accurate predictions, even in non-linear relationships with the help of kernel functions; essentially, it predicts continuous values rather than classifications like a standard SVM does.
- Support Vector Regression (SVR) is a machine learning technique used for regression tasks. It is a variant of Support Vector Machines (SVM) and is designed to predict continuous numeric values, making it suitable for tasks like time series forecasting, stock price prediction, and more.
- SVR is an extension of SVM used for regression problems, where the goal is to predict a continuous value instead of a class label.

## Q8. What is the Kernel Trick in SVM?
- The kernel trick is a technique that allows SVM to handle non-linear data by implicitly mapping the data into a higher dimensional feature space through a kernel function, enabling linear separation in this new space. This is achieved by using a "kernel function" to calculate the similarity between data points without explicitly computing the transformation to the higher-dimensional space, making it computationally efficient.

## Q9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel?
- Linear Kernel: A simple dot product that works well for linearly separable data.
Polynomial Kernel: can capture more complex patterns by mapping data to a higher-dimensional space with polynomial featuresand Introduces non-linearity by raising the features to a power, potentially capturing more complex relationships.
RBF Kernel (Radial Basis Function): A common choice for non-linear data, measuring the similarity between data points based on their distance in feature space.
- often considered the "go-to" kernel for most scenarios where data isn't easily linearly separable; the key difference lies in the type of transformation applied to the data, with linear simply using a dot product, polynomial using polynomial functions, and RBF using a Gaussian distance calculation.


## Q10. What is the effect of the C parameter in SVM?
- In Support Vector Machines (SVM), the C parameter plays a crucial role in controlling the trade-off between maximizing the margin and minimizing classification error.
-  The C parameter is a regularization parameter that helps determine the balance between fitting the model well to the training data and avoiding overfitting.
- A low value of C allows a larger margin, but it can result in more misclassifications (errors) on the training data.
- A high value of C makes the classifier more sensitive to training data, trying to fit the model to correctly classify most of the training points, which can lead to a smaller margin, but fewer misclassifications. This can result in overfitting if the model becomes too focused on the training set.
-When C is small, the model will be more tolerant of misclassification, leading to a simpler, more generalized model.
- When C is large, the model will focus on reducing errors at the cost of a smaller margin, making it more complex and potentially overfitting.
- Choosing the right C value is typically done using techniques like cross-validation to ensure a balance between underfitting and overfitting.

## Q11. What is the role of the Gamma parameter in RBF Kernel SVM?
- The Gamma parameter in a Radial Basis Function (RBF) kernel SVM plays a crucial role in determining the shape and flexibility of the decision boundary. Specifically, gamma controls how far the influence of a single training example reaches.
- When gamma is small, the kernel function will have a longer reach.
It means that each training example will influence a larger region of the feature space.This can lead to underfitting because the model will be too simple and won't capture the complex structure of the data well. The decision boundary will be smoother and less flexible.
- When gamma is large, the influence of each training example will be more localized.Each data point will only have a small region of influence, leading to a more complex and flexible decision boundary.
This can result in overfitting because the model tries to fit the data too closely, capturing noise or small fluctuations in the data as significant patterns.
- In simple terms, small gamma makes the model simpler and large gamma makes it more sensitive to the data, which can make the model too complex.The best gamma value is typically chosen using techniques like cross-validation.

## Q12. What is the Naive Bayes classifier, and why is it called "Naive"?
- The Naive Bayes classifier is a probabilistic machine learning model based on Bayes' Theorem, which is used for classification tasks. It predicts the class of a data point based on the probabilities of the features belonging to each class.
- It is called "naive" because it assumes that all features are independent, even though, in reality, features are often correlated. This assumption is typically not true in practice, but it makes the model easier to compute.
- Despite the naive assumption of independence, Naive Bayes can perform surprisingly well in many situations, especially with text classification tasks (like spam detection) where the independence assumption is often reasonable.
In summary, the Naive Bayes classifier is a simple, fast, and effective probabilistic classifier based on Bayes' Theorem with the naive assumption that features are independent.

## Q13. What is Bayes’ Theorem?
- Bayes' Theorem is a mathematical formula in probability theory that allows you to calculate the conditional probability of an event based on prior knowledge of related events, essentially updating your belief about an event as new evidence becomes available; it essentially calculates the probability of "A" given that "B" has occurred, using the prior probabilities of "A" and "B" and the likelihood of "B" given "A" as key factors.
- Formula:$ P(A|B) = (P(B|A) * P(A)) / P(B) $

 P(A|B): Probability of event A occurring given that event B has occurred (posterior probability)

 P(B|A): Probability of event B occurring given that event A has occurred (likelihood)

 P(A): Prior probability of event A

 P(B): Probability of event B occurring (marginal probability)

- Applications:
Medical diagnosis: Calculating the probability of a disease given a positive test result
Machine learning: Updating predictions based on new data
Data analysis: Inferring hidden patterns from observed data
## Q14. Explain the differences between Gaussian Naive Bayes, Multinomial Naive Bayes, and Bernoulli Naive Bayes?
- Here’s a quick breakdown of the three types of Naive Bayes:

1. Gaussian Naive Bayes:
   - Data Type: Continuous data (e.g., height, weight).
   - Assumption: Features follow a normal distribution.
   - Use Case: For problems with continuous features.

2. Multinomial Naive Bayes:
   - Data Type: Count data (e.g., word frequencies).
   - Assumption: Features follow a multinomial distribution.
   - Use Case: Common in text classification (word counts).

3. Bernoulli Naive Bayes:
   - Data Type: Binary data (e.g., presence or absence of a feature).
   - Assumption: Features follow a Bernoulli distribution (0 or 1).
   - Use Case: For binary features (like yes/no).

## Q16. What are the key assumptions made by Naive Bayes?
- The key assumptions made by Naive Bayes are:
 1. Naive Bayes assumes that all features (variables) are independent of each other, given the class label. This means that the presence or value of one feature does not affect the presence or value of another feature, which is often unrealistic in real-world data, but simplifies calculations.
 2. It assumes that the probability of a feature belonging to a certain class depends only on the class itself and not on other features, leading to the "naive" assumption of feature independence.
- Independence of features given the class label.
- Conditional probability of each feature given the class.

 4. It may perform poorly if the dataset is highly imbalanced, as it tends to favor the majority class.

## Q20 How does Laplace Smoothing help in Naive Bayes?
- Laplace Smoothing (also called Additive Smoothing) helps in Naive Bayes by handling the problem of zero probability for features that don't appear in the training data for a given class.

- In Naive Bayes, we calculate probabilities of features (like words) for each class. If a feature doesn’t appear in the training data for a class, its probability would be zero, which can cause issues when multiplying probabilities for prediction.
- Laplace Smoothing adds a small constant (usually 1) to each feature count to ensure that no probability is zero.

- Formula:
$[
P(w|C) = \frac{Count(w, C) + 1}{Count(C) + V}
]$

Where:
- Count(w, C): Count of word w in class C.
- Count(C): Total count of words in classC.
- V: Total number of unique features (e.g., words in the dataset).


- Prevents zero probabilities, ensuring more reliable predictions.
- Helps with sparse data (when some features don’t appear in certain classes).

- Laplace Smoothing ensures that all features have a non-zero probability, improving the robustness of the Naive Bayes classifier.

## Q38. Write a Python program to train an SVM Classifier using Stratified K-Fold Cross-Validation and compute the average accuracy?
### Ans

```python
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import accuracy_score

data = load_digits()
df = pd.DataFrame(data.data)
df['target']= data.target

# x and y
x = df.drop('target',axis=1)
y = df['target']

#train test split
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=42)

#cross validation
skf = StratifiedKFold(n_splits=5,shuffle=True,random_state=42)
accuracies = []
for train_index,test_index in skf.split(x,y):
    x_train,x_test = x.iloc[train_index],x.iloc[test_index]
    y_train,y_test = y.iloc[train_index],y.iloc[test_index]
    model = SVC(kernel='linear')
    model.fit(x_train,y_train)
    y_pred = model.predict(x_test)
    accuracy = accuracy_score(y_test,y_pred)
    accuracies.append(accuracy)

#average accuracy
average_accuracy = np.mean(accuracies)
print(f"Average accuracy: {average_accuracy}")
```
