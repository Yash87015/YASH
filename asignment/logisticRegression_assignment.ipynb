## Q1. What is Logistic Regression, and how does it differ from Linear Regression?
- Logistic Regression is a classification algorithm that predicts probabilities using the sigmoid function. Unlike Linear Regression, which predicts continuous values, Logistic Regression is used for categorical outcomes.

## Q2 What is the mathematical equation of Logistic Regression?
- $P(Y=1) = 1 / (1 + e^-(b0 + b1X1 + b2X2 + ... + bnXn))$

- where β₀ is the intercept, β₁ to βₙ are coefficients for each feature X₁, X₂, ..., Xₙ, and "e" is the base of the natural logarithm.

## Q3. 3. Why do we use the Sigmoid function in Logistic Regression?
- The Sigmoid function converts any real-valued input into a probability range (0 to 1), making it suitable for classification.
- sigmoid function below
 $$ f(x) = \frac{1}{1 + e^{-x}} $$

## Q4. What is the cost function of Logistic Regression?
- Logistic Regression uses the log loss (binary cross-entropy) as the cost function:
- $J(θ) = -(1/m) ∑ [ yi log(ŷi) + (1 - yi) log(1 - ŷi) ]$

$$ J(\theta) = \frac{1}{m} \sum_{i=1}^{m} \left[ -y^{(i)} \log(h_{\theta}(x^{(i)})) - (1 - y^{(i)}) \log(1 - h_{\theta}(x^{(i)})) \right] $$

- where
1. J(θ) is the cost function.
2. m is the number of training examples.
3. (yi) is the true label of the i-th training example (either 0 or 1).
4.  ŷi:-is the predicted probability (using the sigmoid function) for the i-th training example.
5.  $h_{\theta}x^{(i)}$ is the hypothesis function (sigmoid function)

## Q5. What is Regularization in Logistic Regression? Why is it needed?
- Regularization (L1/L2/elastic net) prevents overfitting by adding a penalty term to the loss function, controlling model complexity.
- You can choose one or combine both (Elastic Net) depending on your specific needs
- L1 regularization can drive some of the model's coefficients  to exactly zero, effectively performing feature selection (removing irrelevant features). It's useful when you suspect that only a small subset of features are important for the model.
- L2 regularization shrinks the model parameters towards zero but does not set them exactly to zero. It prevents overfitting by penalizing large coefficients, making the model simpler and more stable.

## Q6. Explain the difference between Lasso, Ridge, and Elastic Net regression.
- Lasso (L1): Shrinks coefficients and can set some to zero (feature selection).
- Ridge (L2): Shrinks coefficients but does not eliminate them.
- Elastic Net: Combination of L1 and L2 regularization.

## Q7. When should we use Elastic Net instead of Lasso or Ridge?
- When features are highly correlated, Elastic Net is preferred as it combines the benefits of both L1 and L2.

## Q8. What is the impact of the regularization parameter (λ) in Logistic Regression?
- Impact on coefficients:
A higher λ value results in smaller coefficients, effectively reducing the influence of individual features on the prediction.
-Bias-Variance Tradeoff:
Increasing λ increases the model's bias (making it simpler) but decreases its variance (reducing overfitting).
-Overfitting Prevention:
By penalizing large coefficients, regularization helps prevent the model from learning noise in the training data, improving its ability to generalize to new data.

- High λ: More regularization, simpler model (risk of underfitting) failing to capture important patterns.

- Low λ: Less regularization, complex model (risk of overfitting) leading to poor performance on new data.

- -Choosing the right λ:
The optimal value of λ is typically determined using cross-validation, where different λ values are tested to find the one that provides the best performance on unseen data.

## Q9. What are the key assumptions of Logistic Regression?

- Recall:- The ratio of key classes identified by an approach to the total number of true key classes
- F1-score:-The harmonic mean of precision and recall, which provides a balanced evaluation
- ROC-AUC:-A statistical metric that measures a model's ability to classify instances from different classes correctly

## Q12. How does class imbalance affect Logistic Regression?
- When dealing with class imbalance in a dataset, Logistic Regression tends to become biased towards the majority class, leading to poor predictive performance for the minority class, especially when trying to identify rare events; this is because the model prioritizes minimizing overall errors, which are often dominated by the majority class, resulting in inaccurate predictions for the less frequent class.
-It can cause biased predictions; solutions include oversampling, undersampling, or using weighted loss functions.

**How to address class imbalance in Logistic Regression:**
- Data sampling techniques:
Oversampling: Replicating data points from the minority class to create a more balanced distribution.
- Undersampling: Removing data points from the majority class to reduce its size.
- Cost-sensitive learning:
Assigning higher weights to misclassifications of the minority class during training to penalize errors more heavily.
- Choosing appropriate evaluation metrics:
Instead of relying solely on overall accuracy, use metrics like precision, recall, F1-score, or AUC-ROC that are more sensitive to minority class performance.

## Q13. What is Hyperparameter Tuning in Logistic Regression?
- Hyperparameter tuning in Logistic Regression refers to the process of adjusting specific settings (called hyperparameters) within the model to optimize its performance by finding the best values for parameters that control the learning process, like regularization strength, solver type, and penalty, which are not learned directly from the data during training, but instead need to be manually set to achieve the best results on a given dataset.
-It involves optimizing parameters like regularization strength (C), solver type, and penalty to improve model performance.
- Purpose:
To improve the accuracy and generalization ability of the Logistic Regression model by finding the most suitable hyperparameter values.
- Common hyperparameters to tune:
1. Regularization strength (C): Controls how much the model penalizes large coefficients, preventing overfitting.
2. Solver: The optimization algorithm used to fit the model, such as 'liblinear', 'lbfgs', or 'saga'.
3. Penalty: The type of regularization applied, like 'l1' (Lasso) or 'l2' (Ridge).
- Tuning methods:
1. Grid Search: Trying out all possible combinations of hyperparameter values within a specified range.
2. Random Search: Randomly sampling combinations of hyperparameter values.
3. Bayesian Optimization: Using a probabilistic approach to efficiently explore the search space.

## Q15. How is Logistic Regression extended for multiclass classification?
- Using One-vs-Rest (OvR) or Softmax Regression (Multinomial Logistic Regression).
- To extend logistic regression for multiclass classification, the most common approach is to use "multinomial logistic regression," which replaces the standard logistic function with a softmax function, allowing the model to calculate probabilities for each class simultaneously, ensuring that the probabilities across all classes sum to 1; this method is considered a more direct extension of binary logistic regression to handle multiple classes.
Key points about extending logistic regression for multiclass classification:
- Softmax function:
The softmax function is crucial in multinomial logistic regression, as it transforms the raw output (logits) into a probability distribution across all classes.
- One-vs-Rest (OvR) strategy:
Another approach is to train multiple binary logistic regression models, where each model compares one class against all other classes combined.
- Choosing the method:
Depending on the specific problem and dataset, either the multinomial logistic regression (preferred for balanced classes) or the one-vs-rest approach can be used.

## Q18. What is the difference between Softmax Regression and Logistic Regression?
- The key difference between Softmax Regression and Logistic Regression is that Softmax Regression is used for multi-class classification problems, where you need to predict one of several possible categories, while Logistic Regression is specifically designed for binary classification tasks, meaning it can only predict between two possible outcomes (like yes/no or spam/not spam).
- Logistic Regression: Binary classification.
- Softmax Regression: Multiclass classification by assigning probabilities across multiple classes.

## Q19. How do we choose between One-vs-Rest (OvR) and Softmax for multiclass classification?
- OvR:OvR is preferable if you have a large number of classes and computational efficiency is a primary concern , trains separate classifiers for each class.
  -  OvR is simpler and faster but may not accurately capture class relationships when there are significant class imbalances.
- Softmax:select Softmax if you need to directly model the probability of each class relative to the others, especially when dealing with well-balanced classes Preferred when classes are mutually exclusive
  - essentially, Softmax provides a more nuanced probability distribution across all classes


- When to use OvR:

1. Large number of classes:
When dealing with a large number of classes, training separate binary classifiers for each class with OvR can be computationally efficient.
2. Interpretability:
If understanding the decision-making process for each class individually is important, OvR can be easier to interpret.

- When to use Softmax:

1. Balanced classes:
When all classes have roughly equal representation in your data, Softmax provides a more accurate probability distribution across classes.
2. Complex class relationships:
If there are intricate relationships between classes where the probability of one class depends on the probabilities of others, Softmax can capture these better.

## Q20. How do we interpret coefficients in Logistic Regression?
- In Logistic Regression, coefficients represent the change in the "log odds" of the outcome variable for a one-unit increase in the corresponding predictor variable, holding all other variables constant; to interpret them more intuitively, you typically exponentiate the coefficient to get the "odds ratio", which tells you how much the odds of the event increase or decrease for a one-unit change in the predictor variable.
- Coefficients represent the change in log-odds for a one-unit change in the predictor. Higher coefficients mean a stronger impact on the outcome.

  - Sign of the coefficient:
A positive coefficient indicates that increasing the predictor variable increases the odds of the event occurring, while a negative coefficient indicates the opposite.

  - Odds Ratio:
To get a more interpretable measure, exponentiate the coefficient to obtain the odds ratio. An odds ratio greater than 1 means the event is more likely to occur with a one-unit increase in the predictor.
  - Interpretation for categorical variables:
When dealing with categorical variables, the coefficient is interpreted relative to the reference level.

```python
#define function to check all columns otlier is preset if present then retirn coulmn name
def check_outlier(df):
    for col in df.columns:
      if df[col].dtype != 'object':  # Check if the column is numerical
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR
        if df[(df[col] < lower_bound) | (df[col] > upper_bound)].shape[0] > 0 :
          return col
    return None # Return None if no outliers are found in numerical columns
```

## Q24. Write a Python program to train Logistic Regression and find the optimal C (regularization strength) using cross-validation?
### Ans

```python
df = sns.load_dataset('titanic')
x = df.drop('survived', axis=1)
y = df[['survived']]

# Preprocessing (handling missing values and categorical features)
cat_cols = ['sex', 'embarked', 'class', 'who', 'adult_male', 'deck', 'embark_town', 'alive', 'alone']
num_cols = ['pclass', 'age', 'sibsp', 'parch', 'fare']

cat_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='most_frequent')),
    ('onehot', OneHotEncoder(handle_unknown='ignore'))
])
num_transformer = Pipeline(steps=[
    ('imputer', SimpleImputer(strategy='median')),
    ('scaler', StandardScaler())
])

preprocessor = ColumnTransformer(
    transformers=[
        ('cat', cat_transformer, cat_cols),
        ('num', num_transformer, num_cols)
    ])


pipeline = Pipeline(steps=[('preprocessor', preprocessor),
                           ('classifier', LogisticRegression())])


# Hyperparameter tuning using GridSearchCV
param_grid = {'classifier__C': [0.001, 0.01, 0.1, 1, 10, 100],
              'classifier__solver': ['liblinear', 'saga', 'newton-cg', 'lbfgs'],
             }
grid_search = GridSearchCV(pipeline, param_grid, cv=5, scoring='accuracy', verbose=3)

x_train, x_test, y_train, y_test = train_test_split(x,y,test_size=0.2,random_state=1)

grid_search.fit(x_train, y_train)


print("Best C:", grid_search.best_params_['classifier__C'])
print("Best solver:", grid_search.best_params_['classifier__solver'])
print("Best Accuracy",grid_search.best_score_)


# Evaluate the best model
best_model = grid_search.best_estimator_
y_pred = best_model.predict(x_test)
accuracy = accuracy_score(y_test, y_pred)
print("Test Accuracy:", accuracy)
```
