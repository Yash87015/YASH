merge:-In Pandas, the merge() function is used to combine two or more DataFrames or named Series objects based on common columns or indices. It performs a database-style join operation, similar to SQL joins, allowing you to align and combine data from different sources.

1) left: The first DataFrame or named Series to merge.
2) right: The second DataFrame or named Series to merge.
3) how: The type of join to perform. Options are:
4) 'inner': Keep only rows that have matching values in both DataFrames (default).
5) 'left': Keep all rows from the left DataFrame and add matching rows from the right DataFrame. 
 
6) 'right': Keep all rows from the right DataFrame and add matching rows from the left DataFrame.
7) 'outer': Keep all rows from both DataFrames, filling in missing values with NaN.
8) 'cross': Create the Cartesian product of both DataFrames (all possible combinations of rows).
9) on: The column or index level name(s) to join on. If not specified, the function tries to infer the common column(s).
10) left_on: The column(s) in the left DataFrame to use as join keys.
11) right_on: The column(s) in the right DataFrame to use as join keys.
12) left_index: Use the left DataFrame's index as join keys.
13) right_index: Use the right DataFrame's index as join keys.
14) suffixes: A tuple of strings to append to overlapping column names in the merged DataFrame.

```python
d3=pd.DataFrame({'key1':[1,2,3,4,5],'key2':[6,7,8,9,10],'key3':[11,12,13,14,15]})
```

```python
d4=pd.DataFrame({'key1':[1,18,19,4,5],'key4':[21,8,9,24,25],'key5':[26,13,14,29,30]})
```

```python
d4
```

```python
pd.merge(d3,d4)#'inner': Keep only rows that have matching values in both DataFrames (default)
```

```python
pd.merge(d3,d4,how='left')#left: The first DataFrame or named Series to merge.
```

```python
pd.merge(d3,d4,how='right')#right: The second DataFrame or named Series to merge.
```

```python
pd.merge(d3,d4,how='outer')#'outer': Keep all rows from both DataFrames, filling in missing values with NaN.
#all the common element of the common column from both the dataframe
```

```python
pd.merge(d3,d4,how='cross')#'cross': Create the Cartesian product of both DataFrames (all possible combinations of rows).
```

```python
#left_on: The column(s) in the left DataFrame to use as join keys.
#right_on: The column(s) in the right DataFrame to use as join keys.
pd.merge(d3,d4,how='left',left_on='key2',right_on="key4")
```

differnce between join and merge:-join works based on index and merge works based on columns

â†’ In Pandas, the join() function is used to combine two DataFrames based on their indexes. It's a convenient way to merge DataFrames when you want to align them based on their index values rather than specific columns.

â†’ Definition:-The join() method combines the columns of two or more DataFrames based on their index values. By default, it performs a left join, meaning all rows from the left DataFrame are kept, and matching rows from the right DataFrame are added.

Key parameters:
1) other: The DataFrame or Series to join with.
2) on: Column or index level name(s) in the caller to join on the index in other. 
3) how: The type of join to perform ('left', 'right', 'inner', 'outer').
4) lsuffix: Suffix to add to overlapping column names from the left DataFrame.
5) rsuffix: Suffix to add to overlapping column names from the right DataFrame.
6) sort: Sort the resulting DataFrame by the join key.

```python
#another way
import time
start = time.time()
for i in range(1000):
    pass
end = time.time()
print("Time taken for the execution of code", end-start)
```

```python
#one more way #best way
start = time.perf_counter()
for i in range(1000):
    pass
end = time.perf_counter()
print("Time taken for the execution of code in seconds", end-start)
```
